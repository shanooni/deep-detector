```python
import numpy as np
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
```

```python
inputs = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/efficientnet_b0_features.npy")
labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/efficientnet_b0_labels.npy")
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, labels, test_size=0.25, random_state=42)
```

```python
classifier = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = classifier.fit(X_train, X_test, y_train, y_test)
```

```python
print("\nðŸ Model Comparison Results:")
print(models)
```

```python
# Save the model performance summary
models.to_csv("outputs/efficientnet_b0_lazyclassifier_results.csv", index=True)
print("âœ… Results saved to outputs/efficientnet_b0_lazyclassifier_results.csv")
```
