```python
import numpy as np
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif
```

```python
## Load features and labels
```

```python
resnet_18_features = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet_features.npy")
resnet_18_labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet_labels.npy")
```

```python
selector = SelectKBest(f_classif, k=512)
resnet18_kbest_sel = selector.fit_transform(resnet_18_features, resnet_18_labels)
```

```python
resnet_18_X_train, resnet_18_X_test, resnet_18_y_train, resnet_18_y_test = train_test_split(resnet18_kbest_sel, resnet_18_labels, test_size=0.25, random_state=42)
```

```python
classifier = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
resnet18_models, resnet18_predictions = classifier.fit(resnet_18_X_train, resnet_18_X_test, resnet_18_y_train, resnet_18_y_test)
```

```python
resnet18_models.to_csv("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/outputs/resnet18_kbest_lazyclassifier_results.csv", index=True)
```
