```python
import numpy as np
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
inputs = np.load("features/efficientnet_b4_features.npy")
labels = np.load("features/efficientnet_b4_labels.npy")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    inputs, labels,
    test_size=0.25,
    random_state=42
)
```

```python
lazy_classifier = LazyClassifier(verbose=1, ignore_warnings=True, custom_metric=None)
```

```python
models, predictions = lazy_classifier.fit(X_train, X_test, y_train, y_test)
```

```python
models.to_csv("data/efficientnet_b4_lazyclassifier_results.csv", index=True)
```
