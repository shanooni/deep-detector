```python
resnet_18_features = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet_features.npy")
resnet_18_labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet_labels.npy")

resnet_50_features = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet50_features.npy")
resnet_50_labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet50_labels.npy")

resnet_34_features = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet34_features.npy")
resnet_34_labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet34_labels.npy")

resnet_101_features = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet101_features.npy")
resnet_101_labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet101_labels.npy")

resnet_152_features = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet152_features.npy")
resnet_152_labels = np.load("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/extracted_features/resnet152_labels.npy")
```

```python
selector = VarianceThreshold(threshold=0.01)
resnet_18_feature_selected = selector.fit_transform(resnet_18_features)
```

```python
resnet_18_X_train, resnet_18_X_test, resnet_18_y_train, resnet_18_y_test = train_test_split(resnet_18_feature_selected, resnet_18_labels, test_size=0.25, random_state=42)
```

```python
classifier = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
resnet18_models, resnet18_predictions = classifier.fit(resnet_18_X_train, resnet_18_X_test, resnet_18_y_train, resnet_18_y_test)
```

```python
resnet18_models.to_csv("/Users/shanoonissaka/Documents/school/thesis-project/code/image-detector/outputs/resnet18_filter_lazyclassifier_results.csv", index=True)
```
